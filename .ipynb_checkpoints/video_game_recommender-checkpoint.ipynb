```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for x in text.split():
        y.append(ps.stem(x))
        
    return " ".join(y)
```

# TfidfVectorizer
tf-idf jest to metoda polegająca na skalowaniu cech tekstowych, która jest używana do określenia ważności słów w danym ###dokumencie w odniesieniu do całego zbioru dokumentów. Wartości tf-idf są obliczane jako iloczyn częstotliwości wystąpienia danego słowa w dokumencie (tf) i odwrotności częstotliwości wystąpienia tego samego słowa w całym zbiorze dokumentów (idf).

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfvectorizer = TfidfVectorizer(stop_words='english',  ngram_range=(2, 2), min_df=6)
```

```python
vectorizer  = tfidfvectorizer.fit_transform(new_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import euclidean_distances
```

```python
euclidean_similarity = euclidean_distances(vectorizer)
```

```python
def recommend(game):
    index = new_df[new_df['game_title2'] == game].index[0]
    distances = sorted(list(enumerate(euclidean_similarity[index])),reverse=False,key = lambda x: x[1])[1:6]
    
    for i in distances:
        print(new_df.iloc[i[0]].game_title2)
       
```

```python
recommend('God of War')
```
